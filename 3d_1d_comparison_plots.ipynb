# Comparing Checlair 1d spectral model of Earth to 3d model with GlobES

### resolving power @ 140, Geometric albedo plots, NMAX=4, LMAX=60, 3d_binning= 6 (auto set minimum in GlobES)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
e0_3d= np.genfromtxt('./true_psg_files/psg_rad_3d_e0.txt') # 216 calculations, binning=5
e0_1d= np.genfromtxt('./true_psg_files/psg_rad_1d_e0.txt') # 1 calculation
# binning=10
e0_3d_10= np.genfromtxt('./true_psg_files/psg_rad_e0_10bin.txt') #70 calculations
#binning=15
e0_3d_15= np.genfromtxt('./true_psg_files/psg_rad_e0_15bin.txt') #24 calculations
# binning= 20
e0_3d_20= np.genfromtxt('./true_psg_files/psg_rad_1d_e0_20bin.txt') #12 calculations
```

```python
plt.figure(figsize=(14,8), dpi=180)
plt.plot(e0_1d[:,0], e0_1d[:,1], color='red', label='1-D Model')
plt.plot(e0_3d_20[:,0], e0_3d_20[:,1], color='gold', label='3-D Model: 12 calculations')
plt.plot(e0_3d_15[:,0], e0_3d_15[:,1], color='limegreen', label='3-D Model: 24 calculations')
plt.plot(e0_3d_10[:,0], e0_3d_10[:,1], color='blue', label='3-D Model: 70 calculations')
plt.plot(e0_3d[:,0], e0_3d[:,1], color='violet', label='3-D Model: 216 calculations (PSG max)')
plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Geometric Albedo')
plt.grid()
plt.legend()
plt.title('Resolving Power=140')
plt.suptitle('3-D VS 1-D Modern Earth Simulated Spectra Comparison', size=20);
#plt.savefig('3d_1d_earth_spectral_variation.png')
```

```python
plt.figure(figsize=(14,10), dpi=180)
plt.plot(e0_1d[:,0], e0_1d[:,1], color='red', label='1-D Model')
plt.plot(e0_3d_20[:,0], e0_3d_20[:,1], color='gold', label='3-D Model: 12 calculations')
plt.plot(e0_3d_15[:,0], e0_3d_15[:,1], color='limegreen', label='3-D Model: 24 calculations')
plt.plot(e0_3d_10[:,0], e0_3d_10[:,1], color='blue', label='3-D Model: 70 calculations')
plt.plot(e0_3d[:,0], e0_3d[:,1], color='violet', label='3-D Model: 216 calculations (PSG max)')
plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Geometric Albedo')
plt.ylim(ymin=0.17, ymax=0.28)
plt.grid()
plt.legend()
plt.title('Resolving Power=140')
plt.suptitle('Zoomed Look: 3-D VS 1-D Modern Earth Simulated Spectra Comparison', size=20);
#plt.savefig('zoomed_in_3d_1d_earth_spectral_variation.png');
```

## Check percent difference

```python
comp_mags= (e0_1d[:,1],e0_3d[:,1])
list_test= comp_mags[1]- comp_mags[0]
pd.DataFrame(list_test).describe()
```

# Less than 1% difference between best 3d plot & 1d modeling

```python
np.mean(list_test)*100
```

# Comparing 500 resolving power 

```python
psg_rad_e0_3d_500= np.genfromtxt('./true_psg_files/psg_rad_e0_3d_500.txt')
psg_rad_e0_1d_500= np.genfromtxt('./true_psg_files/psg_rad_e0_1d_500.txt')
```

```python
plt.figure(figsize=(14,8), dpi=220)

plt.plot(psg_rad_e0_3d_500[:,0], psg_rad_e0_3d_500[:,1], color='dodgerblue', label='3-D Earth')
plt.plot(psg_rad_e0_1d_500[:,0], psg_rad_e0_1d_500[:,1], color='magenta', label='1-D Earth')
plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Geometric Albedo')
plt.grid()
plt.legend()
plt.suptitle('3-D VS 1-D Modern Earth Simulated Spectra Comparison', size=20)
plt.title('Higher Resolving Power= 500')
#plt.savefig('500_5p_3d_1d_earth_spectral_variation.png');
```

## check % difference

```python
comp_mags= (psg_rad_e0_1d_500[:,1],psg_rad_e0_3d_500[:,1])
list_test= comp_mags[1]- comp_mags[0]
pd.DataFrame(list_test).describe()
```
