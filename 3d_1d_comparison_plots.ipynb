# Comparing Checlair 1d spectral model of Earth to 3d model with GlobES

### resolving power @ 140, Geometric albedo plots, NMAX=4, LMAX=60, 3d_binning= 6 (auto set minimum in GlobES)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
e0_3d= np.genfromtxt('./true_psg_files/psg_rad_3d_e0.txt')
e0_1d= np.genfromtxt('./true_psg_files/psg_rad_1d_e0.txt')
# binning=10
e0_1d_10= np.genfromtxt('./true_psg_files/psg_rad_e0_10bin.txt')
#binning=15
e0_1d_15= np.genfromtxt('./true_psg_files/psg_rad_e0_15bin.txt')
# binning= 20
e0_1d_20= np.genfromtxt('./true_psg_files/psg_rad_1d_e0_20bin.txt')
```

```python
plt.figure(figsize=(14,12), dpi=180)
plt.plot(e0_1d[:,0], e0_1d[:,1], color='red', label='1d Earth')
plt.plot(e0_1d_10[:,0], e0_1d_10[:,1], color='gold', label='1d Earth: Binning=10')
plt.plot(e0_1d_15[:,0], e0_1d_15[:,1], color='limegreen', label='1d Earth: Binning=15')
plt.plot(e0_1d_20[:,0], e0_1d_20[:,1], color='blue', label='1d Earth: Binning=20')
plt.plot(e0_3d[:,0], e0_3d[:,1], color='violet', label='3d Modern Earth: Binning=6 (minimum)')
plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Geometric Albedo')
plt.grid()
plt.legend()
plt.title('3d VS 1d Earth Simulated Spectra Comparison (Resolving Power=140)')
#plt.savefig('3d_1d_earth_spectral_variation.png');
```

```python
plt.figure(figsize=(14,10), dpi=180)
plt.plot(e0_1d[:,0], e0_1d[:,1], color='red', label='1d Earth')
plt.plot(e0_1d_10[:,0], e0_1d_10[:,1], color='gold', label='1d Earth: Binning=10')
plt.plot(e0_1d_15[:,0], e0_1d_15[:,1], color='limegreen', label='1d Earth: Binning=15')
plt.plot(e0_1d_20[:,0], e0_1d_20[:,1], color='blue', label='1d Earth: Binning=20')
plt.plot(e0_3d[:,0], e0_3d[:,1], color='violet', label='3d Modern Earth: Binning=6 (minimum)')
plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Geometric Albedo')
plt.ylim(ymin=0.17, ymax=0.3)
plt.grid()
plt.legend()
plt.title('3d VS 1d Earth Simulated Spectra Comparison (Resolving Power=140)')
#plt.savefig('zoomed_in_3d_1d_earth_spectral_variation.png');
```

## Check percent difference

```python
comp_mags= (e0_1d[:,1],e0_3d[:,1])
list_test= comp_mags[1]- comp_mags[0]
pd.DataFrame(list_test).describe()
```

# Less than 1% difference between best 3d plot & 1d modeling

```python
np.mean(list_test)*100
```

# Comparing 500 resolving power 

```python
psg_rad_e0_3d_500= np.genfromtxt('./true_psg_files/psg_rad_e0_3d_500.txt')
psg_rad_e0_1d_500= np.genfromtxt('./true_psg_files/psg_rad_e0_1d_500.txt')
```
